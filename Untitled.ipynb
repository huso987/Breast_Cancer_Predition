```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Veri kmesi
df = pd.read_csv("data.csv")
pd.set_option("display.max_rows",None )
pd.set_option("display.max_columns",len(df.columns))
print(df)
```

```python
print(df.info())
```

```python
# Diagnosis stununun deerine gre saylar
diagnosis_counts = df["diagnosis"].value_counts()

# Dairesel grafik Dalm M(iyi huylu tmr) B(Kt huylu tmr)
plt.pie(diagnosis_counts, labels=diagnosis_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  
plt.title("Diagnosis Dalm")
plt.show()
```

```python
# O lar 0 , I lar 1 yapmak iin float64 olanlar Object tibine dntrdm
for column in df.columns:
    if df[column].dtype == "float64":
       df[column] = df[column].astype("object")
print(df.info())
```

```python
# O lar 0 , I lar 1 yapma
for column in df.columns:
    df[column] = df[column].replace({'O': '0', 'I': '1'}, regex=True)
print(df)
```

```python
# (",") leri (".") yapma
def replace_commas_with_dots(df):
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].str.replace(",", ".")
    return df


df = replace_commas_with_dots(df.copy())
print(df)
```

```python
#dignosis ve id stunu hari floata evirdim
for i in range(2,32):
    df[df.columns[i]] = pd.to_numeric(df[df.columns[i]], errors='coerce')
print(df.info())
```

```python
# Sadece saysal stunlar se
numerical_columns = df.select_dtypes(include=['float64']).columns

# Saysal stunlar kullanarak gruplama ve ortalama hesaplama
means_by_diagnosis = df.groupby("diagnosis")[numerical_columns].mean()

# Ortalama deerlere gre ubuk grafik izimi
means_by_diagnosis.plot(kind='bar', figsize=(12, 8), colormap='viridis', edgecolor='black')
plt.title('Kt ve yi Huylu Tmrler in Stun Ortalamalar')
plt.xlabel('Tmr Tehisi')
plt.ylabel('Ortalama Deerler')
plt.xticks(rotation=0)
plt.legend(title='Stunlar', bbox_to_anchor=(1, 1)) 
plt.tight_layout()
plt.show()
```

```python
# NaN deererin says stun bazl
missing_values = df.isna().sum()
missing_values.plot(kind='bar', figsize=(10, 6), color='skyblue', edgecolor='black')
plt.title('Bo Deer Saylar - Her Stun in')
plt.xlabel('Stunlar')
plt.ylabel('Bo Deer Says')
plt.xticks(rotation=45, ha='right')  # Stun etiketlerini dndrme
plt.tight_layout()
plt.show()
```

```python
selected_columns = [
    "perimeter_mean",
    "area_mean",
    "concave points_mean",
    "symmetry_mean",
    "perimeter_se",
    "concave points_se",
    "fractal_dimension_se",
    "radius_worst",
    "area_worst",
    "compactness_worst",
    "concave points_worst",
    "fractal_dimension_worst",
]

# Kt huylu iyi huylu tmre gre ortalama hesaplama
means_by_diagnosis = df.groupby("diagnosis")[selected_columns].mean()
print(means_by_diagnosis)
```

```python
means_by_diagnosis.plot(kind='bar', figsize=(12, 8), colormap='viridis', edgecolor='black')
plt.title('Kt ve yi Huylu Tmrler in Stun Ortalamalar')
plt.xlabel('Tmr Tehisi')
plt.ylabel('Ortalama Deerler')
plt.xticks(rotation=0)
plt.legend(title='Stunlar', bbox_to_anchor=(1, 1)) 
plt.tight_layout()
plt.show()
```

```python
# NaN deerleri doldur
for diagnosis in ["B", "M"]:
    for column in selected_columns:
        # NaN deerlerini, ilgili tehis ve stunun ortalamasyla doldur
        df.loc[df["diagnosis"] == diagnosis, column] = df.loc[df["diagnosis"] == diagnosis, column].fillna(means_by_diagnosis.loc[diagnosis, column])

# Kontrol iin NaN deerleri say
nan_counts_after_fillna = df[selected_columns].isna().sum()

# ilemden gemi data
print(df)
```
